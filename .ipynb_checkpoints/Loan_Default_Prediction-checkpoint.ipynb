## Taking Care of Missing Values

### Categorical Missing Values

```python
columns = ['loan_limit', 'approv_in_adv', 'loan_purpose', 'Neg_ammortization', 'age', 'submission_of_application']
```

```python
df['loan_limit'].mode()[0]
```

```python
# function to fill the columns with categorical empty values in our dataset with the most occurring values in the column (Mode)
for i in range(len(columns)):
    df[columns[i]].fillna(df[columns[i]].mode()[0],inplace=True)
```

```python
df.isnull().sum()*100/len(df)
```

### Numerical Missing Values

```python
num_columns = ['rate_of_interest', 'Interest_rate_spread', 'Upfront_charges', 'property_value', 'income', 'LTV', 'dtir1', 'term']
```

```python
df['rate_of_interest'].mean()
```

```python
# function to fill the columns with numerical empty values in our dataset with the average values in the column (Mean)
for i in range(len(num_columns)):
    df[num_columns[i]].fillna(df[num_columns[i]].mean(),inplace=True)
```

```python
df.isnull().sum()*100/len(df)
```

```python
df.head()
```

## Encoding Categorical Values

```python
df.info()
```

```python
# creating a list of columns with object data types
categorical_columns = list(df.select_dtypes(include=['object']).columns)

# Print the list of object-type column names
print(categorical_columns)
```

```python
df[categorical_columns[0]]
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform categorical columns
encoded_array = encoder.fit_transform(df[categorical_columns])

# Debugging: Check shape mismatch
# print("Encoded array shape:", encoded_array.shape)
# print("Feature names length:", len(encoder.get_feature_names_out(categorical_columns)))

# Convert encoded array to DataFrame
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_columns))

# Drop original categorical columns and merge with encoded data
df_updated = df.drop(columns=categorical_columns).join(encoded_df)
```

```python
df_updated.sample(5)
```

## Dropping Independent Variable

```python
x = df_updated.drop('Status', axis = 1)
y = df_updated['Status']
```

```python
x = x.values
print(x)
```

```python
x.shape
```

```python
y = y.values
print(y)
```

```python
y.shape
```

## Splitting Data into Trainig and Test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
x_train
```

```python
y_train
```

## Feature Scaling

```python
cols = ['loan_amount','rate_of_interest','Credit_Score', 'Upfront_charges', 'property_value','LTV','dtir1', 'term', 'income']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[cols] = sc.fit_transform(x_train[cols])
x_test[cols] = sc.transform(x_test[cols]) # we didn't use fit_transform here so as to avoid information leakage on the test set
```

## Converting Dataframe into Numpy array to feed into model

```python
x_train = x_train.values
print(x_train)
```

```python
x_test= x_test.values
print(x_test)
```

```python
y_train = y_train.values
print(y_train)
```

```python
y_test = y_test.values
print(y_test)
```

```python
y_test.shape
```

## Training Model on Various Classification Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_classifier = LogisticRegression(random_state = 0, max_iter = 500)
logistic_classifier.fit(x_train, y_train)
```

```python
# making predictions on our model
y_pred = logistic_classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy score -- ', accuracy_score(y_test, y_pred))

```
